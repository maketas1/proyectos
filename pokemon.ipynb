Este dataset tiene 800 filas y 12 columnas
Las columnas Name, Type1 y Type2 son de tipo str, la columna Legendary es de tipo bool y el resto de columnas son de tipo int64

La columna Type1 tiene un total de 18 elementos unicos y son los tipos que existen en pokemon
La columna Type2 tiene un total de 19 elementos unicos uno mas que el Type1 debido a que existen pokemons con un unico tipo
La columna Generation tiene un total de 6 elementos unicos, un numero por generacion de juego
La columna Num contiene el numero de pokedex de ese pokemon pero no es unico por pokemon porque existen variantes de pokemons que usan el mismo numero de registro

Columnas HP, Attack, Defense, SpAtk, SpDef, Speed
La media y la mediana nos indica que la mayoria de datos estan sesgados hacia valores altos
La desviacion standar indica que las estadísticas ofensivas varían más que las defensivas

Tipo de variables:
Numerica continua: HP, Attack, Defense, SpAtk, SpDef, Speed
Categorica nominal: Num, Name, Type1, Type2
Categorica ordinal: Generation



```python
fig, ax = plt.subplots(figsize=(16, 6))

sns.barplot(
    data=df1,
    x='Type1',
    y='Num',
    hue='Type1',
    estimator='mean',
    #errorbar=('ci', 95),
    palette='muted',
    orient='v',
    ax=ax
)
```
