Este dataset tiene 139 y 13 columnas
La columna Trip ID es de tipo int64
Las columnas Duration (days) y Traveler age son de tipo float64

La columna Destination tiene 60 valores unicos, que son las distintas ciudades y paises a los que se puede viajar
La columna Start date apenas tiene valores unicos, lo que almacena son las fechas de inicio del viaje
La columna End date apenas tiene valores unicos, lo que almacena son las fechas de finalizacion del viaje
La columna Traveler name apenas tiene valores unicos, lo que almacena son los nombres de los viajeros
La columna Traveler gender tiene tres valores unicos (male, femanle, NaN), los valores estan sesgados al gender Female
La columna Traveler nationality tiene un total de 42 valores unicos, lo que almacena son las nacionalidades de cada uno de los difernetes viajeros
La columna Accommodation type tiene un total de 9 valores unicos (Hotel, Resort, Villa, Airbnb, Hostel, Riad, Vacation rental, nanm Guesthouse), los valores estan sesgados hacia Hotel
La columna Acomodation cost no tiene muchos valores unicos, almacena los diferentes precios de alojamiento, estos estan en dolares pero hay que formatear debido a que en algunos no sale el simbolo, en otros si sale y en otros pone USD
La columna Transportation type tiene 10 valores unicos (Flight, Train, Plane, Bus, Car rental, Subway, nan, Car, Ferry, Airplane), los valores estan sesgados hacia Plain
La columna Transportation cost no tiene muchos valores unicos, almacena los diferentes precios de transporte, estos estan en dolares pero hay que formatear debido a que en algunos no sale el simbolo, en otros si sale y en otros pone USD

Tipo de varables:
Numerica continua: Trip ID, Duration (days), Traveler age
Categorica nominal: Destination, Traveler name, Traveler gender, Traveler nationality, Accomodation type, Acommodation type, Acomodation cost, Transportation type, Transportation cost
Temporal: Start date, End date


```python
df1_copia = df1.copy()
```

```python
fig, ax = plt.subplots(figsize=(8, 5))

ax.hist(df1_copia['Traveler age'])
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.title('Distribución de edades')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

ax.hist(df1_copia['Duration (days)'])
plt.xlabel('Duración')
plt.ylabel('Frecuencia')
plt.title('Duración de los viajes')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 10))

sns.boxplot(
    data=df1_copia,
    y= "Accommodation cost",
    ax=ax
)
```

```python
fig, ax = plt.subplots(figsize=(5, 10))

sns.boxplot(
    data=df1_copia,
    y= "Transportation cost",
    ax=ax
)
```
